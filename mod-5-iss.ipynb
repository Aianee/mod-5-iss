```python
import json
from datetime import datetime, timedelta
import requests
import matplotlib.pyplot as plt
import time

iss = []

ending_time = datetime.now() + timedelta(minutes=20)

while(ending_time > datetime.now()):

    URL = 'http://api.open-notify.org/iss-now.json'
    req = requests.get(url=URL)
    data = req.json()
    time.sleep(10) 
    iss.append(data)
    
with open('iss_data.json','w') as f:
    json.dump(iss,f)
```

```python
with open('iss_data.json','r') as f:
    iss_json = json.loads(f.read())

latitude = []

for i in iss_json:
    latitude.append(i['iss_position']['latitude'])
    
longitude = []

for i in iss_json:
    longitude.append(i['iss_position']['longitude'])
    
x = latitude 
y = longitude 

fig,axes = plt.subplots(figsize=(50,50))

axes.scatter(x,y)
plt.xticks(rotation=90)
axes.set_title("Path of the International Space Station (ISS)",fontsize=45)
axes.set_xlabel('Latitude',fontsize=30)
axes.set_ylabel('Longitude',fontsize=30)
```
